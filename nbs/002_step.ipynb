## 1. DataFrame contracts

Decorator to give functions dataframe contracts
to give clear indication of column expectations

```python
#| export

from functools import wraps, partial
from typing import Iterable
```

```python
#| export

def step(requires: Iterable[str] = (), provides: Iterable[str] = ()):
    reqs = set(requires); provs = set(provides)
    def deco(fn):
        fn.requires = reqs
        fn.provides = provs
        @wraps(fn)
        def wrapper(df, *args, **kwargs):
            missing = fn.requires - set(df.column_names)
            if missing:
                raise ValueError(f"{fn.__name__} missing columns: {missing}")
            out = fn(df, *args, **kwargs)
            return out
        return wrapper
    return deco
```
